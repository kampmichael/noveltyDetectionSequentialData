```python
import json
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score
import pickle
import warnings
warnings.filterwarnings('ignore')
from sklearn.svm import OneClassSVM 
```

```python
X = pickle.load(open('data/X.pckl','rb'))
X_test = pickle.load(open('data/X_test.pckl','rb'))
y_test = pickle.load(open('data/y_test.pckl','rb'))
Xpo = pickle.load(open('data/Xpo.pckl','rb'))
Xpo_test = pickle.load(open('data/Xpo_test.pckl','rb'))
Xpbow = pickle.load(open('data/Xpbow.pckl','rb'))
Xpbow_test = pickle.load(open('data/Xpbow_test.pckl','rb'))
```

```python
#Bag of Words
paramsAUCbow = {}
scaleFact = 1./(X.shape[1]*X.var())
for gamma in [0.001*scaleFact,0.005*scaleFact,0.01*scaleFact,0.05*scaleFact,0.1*scaleFact,0.5*scaleFact,scaleFact,1.5*scaleFact,2*scaleFact]:
    for nu in [2**i for i in range(-12,-5)]:
        ocsvm = OneClassSVM(gamma=gamma, nu=nu)
        ocsvm.fit(X)
        y_pred = ocsvm.decision_function(X_test)
        auc = roc_auc_score(y_test, y_pred)
        print(gamma, nu, False, auc)
        paramsAUCbow[auc] = [gamma,  nu, False]
        #y_pred_inv = np.ones(y_pred.shape) - y_pred
        #auc_inv = roc_auc_score(y_test, y_pred_inv)
        #print(gamma, nu, True, auc_inv)
        #paramsAUCbow[auc_inv] = [gamma,  nu, True]
maxAucbow = max(paramsAUCbow.keys())
bestParamsAucbow = paramsAUCbow[maxAucbow]
print("Best params: ", bestParamsAucbow, " with AUC: ",maxAucbow)
```

```python
#Patterns Only
paramsAUCpo = {}
scaleFact = 1./(X.shape[1]*X.var())
for gamma in [0.001*scaleFact,0.005*scaleFact,0.01*scaleFact,0.05*scaleFact,0.1*scaleFact,0.5*scaleFact,scaleFact,1.5*scaleFact,2*scaleFact]:
    for nu in [2**i for i in range(-12,-5)]:
        ocsvm = OneClassSVM(gamma=gamma, nu=nu)
        ocsvm.fit(Xpo)
        y_pred = ocsvm.decision_function(Xpo_test)
        auc = roc_auc_score(y_test, y_pred)
        print(gamma, nu, False, auc)
        paramsAUCpo[auc] = [gamma,  nu, False]
        #y_pred_inv = np.ones(y_pred.shape) - y_pred
        #auc_inv = roc_auc_score(y_test, y_pred_inv)
        #print(gamma, nu, True, auc_inv)
        #paramsAUCpo[auc_inv] = [gamma,  nu, True]
maxAucpo = max(paramsAUCpo.keys())
bestParamsAucpo = paramsAUCpo[maxAucpo]
print("Best params: ", bestParamsAucpo, " with AUC: ",maxAucpo)
```

# Further Error Measures

```python
def calcSensAndSpec(cm):
    tn = cm[0,0]
    fn = cm[1,0]
    fp = cm[0,1]
    tp = cm[1,1]
    print(cm)
    TPR = float(tp)/float(tp + fn)
    TNR = float(tn)/float(tn + fp)
    return TPR, TNR
```

```python
# Bag of Words
ocsvm = OneClassSVM(gamma=3.7252692712877524e-08, nu=0.001953125)
ocsvm.fit(X)
y_pred_bin = ocsvm.predict(X_test)*-1
y_pred_bin = np.where(y_pred_bin == 1, y_pred_bin, 0.)
cm = confusion_matrix(y_test, y_pred_bin)
print(calcSensAndSpec(cm))
```
